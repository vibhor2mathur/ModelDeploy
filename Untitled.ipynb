```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
sns.set()
np.printoptions('precision=2')
```

```python
data=pd.read_csv('nigeria_houses_data.csv')
```

```python
data['title'].unique()
```

```python
data.isnull().sum()
```

So no null values present in any of the columns

```python
data_num=[feature for feature in data.columns if data[feature].dtype!='O']
data_num
```

```python
data['bedrooms'].value_counts()
```

```python
from numpy import median
for feature in data_num: 
    if feature not in ['price']:
        sns.lineplot(x=feature,y='price',data=data,estimator=median)
        #plt.ylim(0,10000000000)
        plt.show()
```

```python
data_cat=[feature for feature in data.columns if data[feature].dtype=='O']
data_cat
```

## Categorical data Analysis

```python
from numpy import mean
for feature in data_cat: 
    if feature not in ['price','town']:
        plt.figure(figsize=(20,10))
        sns.barplot(y=feature,x='price',data=data,estimator=median)
        plt.show()
```

```python
state_mean=data.groupby(['state'])['price'].median()
state_mean=state_mean/100000000
state_mean=state_mean.to_dict()
state_mean
```

```python
town_mean=data.groupby(['town'])['price'].median()
town_mean=town_mean/100000000
town_mean=town_mean.to_dict()
town_mean
```

```python
data1=data.copy()
```

```python
data=data1.copy()
```

```python
data['state_mean_ratio']=data['state'].map(state_mean)
data.head()
```

```python
data['town_mean_ratio']=data['town'].map(town_mean)
data.head()
```

```python
title_mean=data.groupby(['title'])['price'].median()
title_mean=title_mean/100000000
title_mean=title_mean.to_dict()
title_mean
```

```python
data['title_mean_ratio']=data['title'].map(title_mean)
data.head()
```

```python
data1=data.copy()
```

```python
data=data1.copy()
```

```python
data=data.drop(columns=['title','town','state'])
```

```python
data.columns
```

```python
columns=['bedrooms', 'bathrooms', 'toilets', 'parking_space',
       'state_mean_ratio', 'town_mean_ratio', 'title_mean_ratio','price']
```

```python
data=data[columns]
```

```python
data.head()
```

```python
X=data.iloc[:,:-1].values
y=data.iloc[:,-1].values
y = y.reshape(len(y),1)
y
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,train_size=0.8,random_state=0)
```

```python
from sklearn.preprocessing import StandardScaler,MinMaxScaler
sx=StandardScaler()
#mm=MinMaxScaler
sy=StandardScaler()
X_train=sx.fit_transform(X_train)
y_train=sy.fit_transform(y_train)
```

```python
from sklearn.svm import SVR
reg=SVR()
reg.fit(X_train,y_train)
```

```python
y_pred=reg.predict(X_test)
```

```python
from sklearn.metrics import r2_score
r2_score(y_test,y_pred)
```
